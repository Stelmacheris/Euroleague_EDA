```python
df['Pos'].unique()
```

```python
df['Pos'].replace('PG',"Guard",inplace = True)
df['Pos'].replace('SG',"Guard",inplace = True)
df['Pos'].replace('F',"Forward",inplace = True)
df['Pos'].replace('G',"Guard",inplace = True)
df['Pos'].replace('SF',"Forward",inplace = True)
df['Pos'].replace('PF',"Forward",inplace = True)
df['Pos'].replace('C',"Center",inplace = True)
df['Pos'].replace('GF',"Forward",inplace = True)
df['Pos'].replace('FC',"Center",inplace = True)
```

```python
df['Draft Status'].unique()
```

```python
import re
drf = []
for data in df['Draft Status']:
    if "Undrafted" in data:
        data = 'Undrafted'
        drf.append(data)
    elif "Eligible" in data:
        data = 'Undrafted'
        drf.append(data)
    else:
        temp = data.split()
        data = str(int(temp[4]) + 30)   if temp[2] == '2' else str(temp[4])
        drf.append(data)
df['Draft Status'] = drf
```

```python
df
```

```python
df = df.replace(r'\n',' ', regex=True) 
df = df.replace(r'\r',' ', regex=True) 
df = df.replace(r'United States','USA', regex=True) 
df = df.replace(r'Republic of the Congo','Congo', regex=True) 
df = df.replace(r'New Zealand','Zealand', regex=True) 
```

```python
df
```

```python
newnat = []
for data in df['Nationality']:
    if ' ' in data:
        newnat.append(data.split()[0])
    else:
        newnat.append(data)
df['Nationality'] = newnat
```
